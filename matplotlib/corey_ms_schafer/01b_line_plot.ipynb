```python
from matplotlib import pyplot as plt
```

```python
# print available styles:
print(plt.style.available)
```

```python
# use predefined line colors, width, markers, plot grid, background, legend:
plt.style.use('fivethirtyeight')
```

```python
ages = [i for i in range(25, 35+1)]
ages
```

```python
dev_sal = [38500, 42000, 46700, 49300, 53200, 56000, 62000, 64900, 67000, 69000, 73000]
pydev_sal = [45000, 49000, 54000, 57000, 63000, 66000, 70000, 70000, 71500, 75000, 83500]
jsdev_sal = [38000, 43500, 47000, 49000, 53000, 56000, 62000, 66000, 69000, 69000, 74500]
dev_sal, pydev_sal, jsdev_sal
```

```python
# short format:
plt.plot(ages, dev_sal, 'k--.', label='all devs')
# long format, color by one symbol:
plt.plot(ages, pydev_sal, color='b', linestyle='-.', marker='o', linewidth=3, label='py devs')
# color by hex code:
plt.plot(ages, jsdev_sal, color='#adad3b', label='js devs')
# dev_sal is hidden behind. plt respects the given order

plt.title('median developer’s salary')
plt.xlabel('Age [yrs]')
plt.ylabel('Salary [USD]')
plt.legend()
plt.grid(True)
# remove any padding:
plt.tight_layout()

# save as png:
plt.savefig(r'pics\01b_line_plot.png')
plt.show()
```
