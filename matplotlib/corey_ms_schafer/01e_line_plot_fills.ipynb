```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('data\salary.csv')
data[:11]
```

```python
ages = data['Age']
dev_sal = data['All_Devs']
py_sal = data['Python']
js_sal = data['JavaScript']
```

```python
plt.plot(ages, dev_sal, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_sal, label='Python')

overall_median = 57287

plt.fill_between(ages, py_sal, dev_sal, where=(py_sal > dev_sal), interpolate=True, alpha=0.25, label='above avg')
plt.fill_between(ages, py_sal, dev_sal, where=(py_sal <= dev_sal), interpolate=True, color='red', alpha=0.25, label='below avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.savefig(r'pics\01e_line_plot_fills.png')
plt.show()
```
