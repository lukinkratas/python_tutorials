```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
```

```python
ages = np.arange(25, 35+1)
ages
```

```python
indexes = np.arange(len(ages))
indexes
```

```python
width = .25
```

```python
dev_sal = [38500, 42000, 46700, 49300, 53200, 56000, 62000, 64900, 67000, 69000, 73000]
pydev_sal = [45000, 49000, 54000, 57000, 63000, 66000, 70000, 70000, 71500, 75000, 83500]
jsdev_sal = [38000, 43500, 47000, 49000, 53000, 56000, 62000, 66000, 69000, 69000, 74500]
dev_sal, pydev_sal, jsdev_sal
```

```python
plt.bar(indexes-width, dev_sal, width=width, color='#444444', label='all devs')
plt.bar(indexes, pydev_sal, width=width, color='b', label='py devs')
plt.bar(indexes+width, jsdev_sal, width=width, color='#adad3b', label='js devs')

plt.legend()
plt.xticks(ticks=indexes, labels=ages)
plt.title('median developer’s salary')
plt.xlabel('Age [yrs]')
plt.ylabel('Salary [USD]')
plt.grid(True)
plt.tight_layout()

plt.savefig(r'pics\02a_bar_chart.png')
plt.show()
```
