```python
from matplotlib import pyplot as plt
import numpy as np
from collections import Counter, OrderedDict
```

```python
plt.style.use('fivethirtyeight')
```

```python
import csv

with open('data\languages.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
```

*or*

```python
import pandas as pd

data = pd.read_csv('data\languages.csv')
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))
```

```python
language_counter.most_common(10)
```

```python
languages = []
popularities = []

for language,popularity in language_counter.most_common(15):
    languages.append(language)
    popularities.append(popularity)
```

```python
languages.reverse()
popularities.reverse()
```

```python
plt.barh(languages, popularities)

plt.title('Most popular programming languages')
plt.xlabel('users')

plt.savefig(r'pics\02b_bar_chart.png')
plt.show()
```
