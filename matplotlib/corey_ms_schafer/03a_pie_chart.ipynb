```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
slices = [120, 80]
labels = ['a', 'b']
```

```python
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})

plt.title('pie')
plt.tight_layout()

plt.savefig(r'pics\03a_pie_chart.png')
plt.show()
```
