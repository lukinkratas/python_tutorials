```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
slices = [59000, 55500, 47500, 36500, 36000, 32000, 27000, 23000, 20500, 18500]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell', 'C#', 'PHP', 'C++', 'Typescript']
colors = ['#e6a83c', '#e1e00b', '#1dc13f', '#916bf9', '#2b9efc', '#012564', '#6b9ca1', '#a6e4fc', '#6b3721', '#6f0378']
explode = [0, 0, 0, 0.2, 0, 0, 0, 0, 0, 0]
```

```python
plt.pie(slices, labels=labels, colors=colors, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})

plt.title('Programming languages')
plt.tight_layout()

plt.savefig(r'pics\03b_pie_chart.png')
plt.show()
```
