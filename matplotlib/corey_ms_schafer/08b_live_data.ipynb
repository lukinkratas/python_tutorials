```python
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
plt.style.use('fivethirtyeight')
```

```python
x_vals = []
y_vals = []

index = count()
```

```python
def animate(i):
    data = pd.read_csv(r'data\rand_gen_data.csv')
    x = data['x_value']
    y1 = data['total_1']
    y2 = data['total_2']

    plt.cla()
    plt.plot(x, y1, label='channel 1')
    plt.plot(x, y2, label='channel 2')
    plt.legend(loc='upper left')
    plt.tight_layout()
```

```python
anim = FuncAnimation(plt.gcf(), animate, interval=500)

plt.tight_layout()
# plt.savefig(r'pics\08b_liveData.png', transparent=True)
plt.show()
```
