```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
data = pd.read_csv(r'data\salary.csv')
data[:11]
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, sharex=True)
#sharex = shared x axes

ax1.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax3.plot(ages, py_salaries, label='Python')
ax3.plot(ages, js_salaries, label='JavaScript')
ax3.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

ax1.legend()
ax1.set_title('Salary by Age')
ax1.set_ylabel('Median Salary [USD]')

ax2.legend()
ax2.set_ylabel('Median Salary [USD]')

ax3.legend()
ax3.set_xlabel('Ages [yrs]')
ax3.set_ylabel('Median Salary [USD]')

plt.tight_layout()
fig.savefig(r'pics\09_subplots.png')
plt.show()
```
