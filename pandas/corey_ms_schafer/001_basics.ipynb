```python
import pandas as pd
import requests
import zipfile
```

```python
r = requests.get(r'https://drive.google.com/uc?export=download&id=1QOmVDpd8hcVYqqUXDXf68UMDWQZP0wQV')

with open('stack_overflow_data_2019.zip', 'wb') as f:
    f.write(r.content)

with zipfile.ZipFile('stack_overflow_data_2019.zip', 'r') as data_zip:
    data_zip.extract('survey_results_public.csv')
    data_zip.extract('survey_results_schema.csv')
```
