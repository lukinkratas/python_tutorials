```python
from PIL import Image, ImageFilter
import os
```

```python
# installation for G colab:
# ! pip install Pillow
```

```python
img = Image.open('meme.jpg')
# to show in new window:
# img.show()
# show in jup ntb:
img
```

```python
img.filename, img.size
```

```python
# resize the image:
size=(100,100)
filename, ext = os.path.splitext(img.filename)
img.thumbnail(size)
# thumbnail modifies the image object
img.save(f'{filename}_50x50{ext}')
img
```

```python
# rotate the image:
img.rotate(90).save(f'{filename}_90{ext}')
img.rotate(90)
```

```python
# black and white colors:
img.convert(mode='L').save(f'{filename}_blk_and_wht{ext}')
img.convert(mode='L')
```

```python
# add blur filter:
img.filter(ImageFilter.GaussianBlur(2)).save(f'{filename}_blk_and_wht{ext}')
img.filter(ImageFilter.GaussianBlur(2))
```
