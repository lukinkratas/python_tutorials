```python
import requests
```

```python
r = requests.get('https://xkcd.com/353/')
print (r)
```

```python
print(help(r))
```

```python
# html source:
print(r.text)
```

```python
r = requests.get('https://imgs.xkcd.com/comics/python.png')
with open('comic.png', 'wb') as f:
    f.write(r.content)
```

```python
print(r)
print(r.status_code)
print(r.ok)
print(r.headers)
```

```python
payload = {'page': 2, 'count': 25}
r = requests.get('https://httpbin.org/get', params=payload)
print(r.text)
```

```python
payload = {'username': 'Lukin', 'password': 'xxx'}
r = requests.post('https://httpbin.org/post', data=payload)
json_dict = r.json()
print(json_dict['form'])
```

```python
# very basic authetication:
r = requests.get('https://httpbin.org/basic-auth/lukin/xxx', auth=('lukin', 'xxx'))
print(r.status_code)
print(r.text)

r = requests.get('https://httpbin.org/basic-auth/lukin/xxx', auth=('gigi', 'xxx'))
print(r.status_code)
print(r.text)
```

```python
# timeout
r = requests.get('https://httpbin.org/delay/2', timeout=3)
print(r.status_code)
print(r.text)

r = requests.get('https://httpbin.org/delay/5', timeout=3)
print(r.status_code)
print(r.text)
```
